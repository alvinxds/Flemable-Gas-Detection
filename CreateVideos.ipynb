```python
thermal = skvideo.io.vread("./gas_methane.mp4",outputdict={"-pix_fmt": "gray"})[:, :, :, 0]
#rgb = skvideo.io.vread("./processed/rgb_vid.mp4",outputdict={"-pix_fmt": "gray"})[:, :, :, 0]
```

```python
cap = cv2.VideoCapture('./gas_methane.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    framenum = cap.get(cv2.CAP_PROP_POS_FRAMES)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('methane',gray)
    
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
    
    if(framenum%100==0):
        print (framenum)

     
cap.release()
cv2.destroyAllWindows()
print (framenum)
```

```python
thermal = skvideo.io.vread("./gas_propane.mp4",outputdict={"-pix_fmt": "gray"})[:, :, :, 0]
cap = cv2.VideoCapture('./gas_propane.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    framenum = cap.get(cv2.CAP_PROP_POS_FRAMES)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('methane',gray)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
    if(framenum%100==0):
        print (framenum)

     
cap.release()
cv2.destroyAllWindows()
print (framenum)
```
